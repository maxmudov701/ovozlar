```python
# 1. KERAKLI KUTUBXONALARNI O'RNATISH
!pip install edge-tts gtts

import torch
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import numpy as np
import edge_tts
import asyncio
import IPython.display as ipd
from gtts import gTTS

# ==========================================
# 4. O'ZBEKCHA OVOZ YARATISH (EDGE-TTS)
# ==========================================
async def make_voice(text):
    voice = "uz-UZ-MadinaNeural" # O'zbek ayol ovozi
    communicate = edge_tts.Communicate(text, voice)
    await communicate.save("output.mp3")
    return ipd.Audio("output.mp3")

# ==========================================
# 5. HAMMASINI ISHGA TUSHIRISH
# ==========================================
print("\n--- 3. Ovoz yaratilmoqda ---")
text = "Assalomu alaykum! Barcha kodlar muvaffaqiyatli ishga tushirildi."
# Colab-da async funksiyani kutish
await make_voice(text)
```

### Qo'shimcha ovoz variantlari (Additional voice options)

```python
# Yana bir ovoz yaratish (Misol: uz-UZ-SardorNeural - O'zbek erkak ovozi)
async def make_another_voice(text, voice_name):
    communicate = edge_tts.Communicate(text, voice_name)
    filename = f"output_{voice_name}.mp3".replace('-', '_') # Fayl nomini to'g'irlash
    await communicate.save(filename)
    return ipd.Audio(filename)

print("\n--- 4. Boshqa ovoz variantlari yaratilmoqda ---")
text_sardor = "Bu Ozotbek ovozi. Xush kelibsiz!"
# O'zbek erkak ovozi
await make_another_voice(text_sardor, "uz-UZ-SardorNeural")
```
